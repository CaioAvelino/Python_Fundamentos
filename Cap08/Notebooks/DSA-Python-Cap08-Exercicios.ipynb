### Exercício 6

```python
# Considere os 3 arrays abaixo
# Retorne o valor do array xarr se o valor for True no array cond. Caso contrário, retorne o valor do array yarr.
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5], dtype=np.float32)
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5], dtype=np.float32)
cond = np.array([True, False, True, True, False])
```

```python
d = zip(xarr, yarr, cond)
list(d)
```

```python
for a,b,c in zip(xarr, yarr, cond):
    print(a, b, c)
```

```python
result = [(a if c else b) for a,b,c in zip(xarr, yarr, cond)]
result
```
