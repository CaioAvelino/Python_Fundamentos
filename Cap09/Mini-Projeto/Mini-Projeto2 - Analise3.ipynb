```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
```

```python
list(df)
```

```python
df2 = pd.DataFrame(df.groupby(['fuelType','gearbox']).price.mean())
df2 = df2.reset_index()
df2
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.set(palette='OrRd',font_scale=3)
sns.barplot(data=df2, x='fuelType', hue='gearbox', y='price')
plt.xlabel('Tipo de Combustível')
plt.ylabel('Preço')
plt.legend(title='Câmbio')
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
```

```python
df3 = df.groupby(['vehicleType','gearbox']).powerPS.mean()
df3 = df3.reset_index()
df3
```

```python
fig2, ax2 = plt.subplots(figsize=(20,15))
sns.set(palette='YlGnBu',font_scale=2)
sns.barplot(data=df3, x = 'vehicleType', hue = 'gearbox', y = 'powerPS')
plt.xlabel('Tipo de Veículo')
plt.ylabel('Potência Média')
plt.legend(title='Câmbio')
plt.show()
```

```python
# Salvando o plot
fig2.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
