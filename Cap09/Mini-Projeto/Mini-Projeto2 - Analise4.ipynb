```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

```python
list(df)
```

```python
# Calcule a média de preço por marca e por veículo
```

```python
df1 = pd.DataFrame(df.groupby(['brand','vehicleType']).price.mean())
df1 = df1.reset_index()
df1.head(5)
```

```python
data1 = df1.pivot("brand","vehicleType", "price")
data1.fillna(0,inplace=True)
data1
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.set(font_scale=1.4)
sns.heatmap(data1, cmap='BuGn', annot=True, fmt=".0f")
plt.xlabel('\nTipo de Veículo', fontsize=25)
plt.ylabel('Marca\n', fontsize=25)
plt.title('\nPreços dos tipos de veículos de cada marca\n', fontsize=30)
plt.show()
```
